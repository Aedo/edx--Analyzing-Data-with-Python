# Normalizando datos del dataframe

```python
#Normalizando los datos
#Metodo 01: Escalado de caracteristicas simples (Divide cada valor por el valor maximo para esa caracteristica, haciendo que 
#los nuevos valores oscilen entre 0 y 1)

#Metodo 02: Min-Max (toma cada valor antiguo - valor minimo de la carcteristica y luego lo divide entre el rango de esa 
#caracteristica. Los nuevos valores resultantes oscilan entre 0 y 1)

#Metodo 03: Puntaje z o Puntaje estandar (Para cada valor se le resta miu, que es el promdeio de la funcion y luego se divide
#por la desviacion estandar sigma. Los valores resultantes rondan el 0 y tipicamente oscilan entre tres negativos y tres 
#positivos, pero pueden ser mayores o menores)

#Ejemplo de normalizacion en la caracteristica/columna "Largo" por el Metodo 01

df["Largo"] = df["Largo"]/df["Largo"].max()
```

```python
df.head(10)
```

```python
df["Largo"].head(10)
```

```python
#Ejemplo de normalizacion en la caracteristica/columna "Ancho" por el Metodo 02

#Mostramos 1ro como se ve la columa sin normalizar
df["Ancho"].head(10)
```

```python
#Normalizando la columna "Ancho" y mostrandola normalizada
df["Ancho"] = (df["Ancho"]-df["Ancho"].min()) / (df["Ancho"].max()-df["Ancho"].min())

df["Ancho"].head(10)
```

```python
#Ejemplo de normalizacion en la caracteristica/columna "Alto" por el Metodo 03

#Mostramos 1ro como se ve la columa sin normalizar
df["Alto"].head(10)
```

```python
#Normalizando la columna "Alto" y mostrandola normalizada
df["Alto"] = (df["Alto"] - df["Alto"].mean()) / df["Alto"].std()

df["Alto"].head(10)
```

```python
#2- Agrupacion (como nos puede ayudar a transformar nuestro conjunto de datos)
#Supongamos que necesitamos saber cualquier relacion entre los diferentes de sistemas de accionamiento: adelante, atras y 
#traccion en las cuatro ruedas, y el precio de los vehiculos?
#De forma tal que sepamos que sistema de manejo agrega mas valor a un vehiculo
#Entonces seria bueno si agruparamos todos los datos por diferentes tipos de ruedas motrices y comparar los resultados de estas
#ruedas motrices diferentes entre si.
#En Pandas podemos usar el metodo agrupar (grupo) y se utiliza en variables categoricas, agrupando los datos en subconjuntos
#segun las diferentes categorias de esa variable (Puede agrupar por una sola variable o puede agrupar por multiples variables
#pasando multiples nombres de variables)

#Ejemplo: digamos que queremos encontrar el precio promedio de vehiculos y observar como difieren entre los diferentes tipos de
#estilos de carrocerias y ruedas motrices.
#Para hacer esto:

#1ro seleccionamos las tres columnas de datos que nos interesan
df_test = df[["Ruedas motrices", "Estilo", "Precio"]]

#Agrupamos los datos reducidos de acuerdo con ruedas motrices y estilo de la carroceria
#Como estamos interesados en saber como difiere el precio promedio en todos los ambitos, podemos tomar la media de cada grupo y
#agregarla al final de linea, agrupando los datos en subcategorias y mostrando solo el precio promedio de cada una de cada 
#subcategoria
df_grp = df_test.groupby(["Ruedas motrices", "Estilo"], as_index=False).mean()
df_grp

#Nota: Podemos entonces, ver que segun nuestros datos, los convertibles de traccion trasera asi como los hardtop de traccion
#trasera tienen el valor mas alto, mientras que los hatchbacks de traccion en las cuatro ruedas tienen el valor mas bajo
```

```python
#Una tabla como la anterior no es facil de leer y tampoco es facil de visualizar... entonces, para que sea mas facil de entender
#transformaremos esta tabla en una tabla dinamica utilizando el metodo pivote.
#En la tabla anterior, tanto las ruedas motrices como el estilo de la carroceria son columnas de escucha. Una tabla dinamica
#tiene una variable mostrada a lolargo de las columnas y la otra mostrada a lo largo de las filas

#Utilizaremos el metodo pivote de pandas, para pivotera la variable de estilo de la carroceria para que se muestre a lo largo de
#las columnas, mientras que las ruedas motrices se mostraran por las filas
df_pivot = df_grp.pivot(index="Ruedas motrices", columns="Estilo")
df_pivot

#Nota: los datos de precios ahora se convierten en una cuadricula rectangular, que es mucho mas facil visualizar... muy parecido
#a lo que generalmente hacemos en hojas excel 
```

```python
#Otra forma de representar la tabla dinamica es a traves de diagramas de mapas de calor...
#El mapa de calor toma una cuadricula rectangular de datos y asigna una intensidad de color basada en el valor de los datos en
#los puntos de la cuadricula... es una excelente manera de trazar la variable objetivo sobre multiples variables y a traves de
#eso obtener pistas visuales con la relacion entre estas variables y el objetivo

#En este ejemplo utilizaremos el metodo pcolor de pyplot para trazar el mapa de calor y convertir la tabla dinamica anterior en
#una forma grafica
#Especificamos el esquema de color rojo-azul
plt.pcolor(df_pivot, cmap="RdBu")
plt.colorbar()
plt.show()
```

```python
#5- Correlacion estadistica (Varios metodos)

#Metodo de Pearson (Nos entregara dos valores: el Coeficiente de correlacion y el valor P)

#Coeficiente de correlacion
#Un valor cercano a 1, implica correlacion positiva
#Un vaor cercano a -1, implica correlacion negativa
#Un valor cercano o = a 0, significa que no existe correlacion

#Valor P
#Un valor P menor que 0.001, nos da una fuerte certeza sobre el coeficiente de correlacion q calculamos
#Un valor P entre 0.001 y 0.05, nos da una certeza moderada
#Un valor P entre 0.05 y 1, nos dara una certeza debil
#Un valor P mayor que 1, no nos dara ninguna certeza de correlacion

#Por tanto, y como ejemplo, podemos decir que existe una FUERTE CORRELACION (Directa y fuerte) cuando el Coeficiente de 
#correlacion es cercano a 1 y el valor P es menor que 0.001

#Calculando la correlacion de Pearson de "Caballos de fuerza" y el "Precio"

#Usando el paquete scipy
from scipy import stats

cor_pearson = stats.pearsonr(df["Caballos de fuerza"], df["Precio"])
cor_pearson

#Nota: Coeficiente de correlacion=0.81 (Cercano a 1, por tanto Correlacion positiva) 
#      y P_Value=5.92e-48 (Es muy pequeño, mas pequeño que 0.001, por tanto Fuerte certeza)

#Por lo que concluimos que la correlacion de los "Caballos de fuerza" y el "Precio" segun Pearson es, Fuerte positiva 
```

```python
#3- Analisis de varianza (ANOVA)
#Supongamos que necesitamos analizar una variable categorica y ver su correlacion entre las diferentes categorias...
#Por ejemplo: en nuestro conjunto de datos del automovil, nos pudiesemos preguntar: de las variables categoricas que tenemos,
#cual tiene impacto en el precio
#Ejemplo: De nuestro conjunto de datos de automovil, podemos usar AMOVA para ver si hay alguna diferencia en el precio promedio
#de marcas como honda y subaru
#ANOVA es una prueba estadistica que significa Analisis de varianza (Se utiliza para encontrar correlacion entre diferentes
#grupos de una variable categorica)
#ANOVA devuelve dos valores, Prueba F, y el Valor P
#Prueba F: Calcula la relacion la variacion entre grupos de media, sobre la variacion de cada uno de los grupos de muestra
#Valor P: Muestra si el resultado obtenido es estadisticamente significativo

#Usando el paquete scipy
from scipy import stats

#1ro, extraemos los datos de "Marca" y "Precio"
df_anova = df[["Marca", "Precio"]]

#2do, agrupamos los datos por diferentes Marcas
group_anova = df_anova.groupby(["Marca"])

#Usando el metodo f_oneway de scipy
resultado_anova = stats.f_oneway(group_anova.get_group("honda")["Precio"], group_anova.get_group("subaru")["Precio"])
resultado_anova

#Nota: los precios de honda y subaru no son significativamente diferentes, ya que el puntaje de la Prueba F es menor que 1 y el
#Valor P es mayor que 0.05
```

```python
#Podemos hacer lo mismo para honda y jaguar

#1ro, extraemos los datos de "Marca" y "Precio"
df_anova01 = df[["Marca", "Precio"]]

#2do, agrupamos los datos por diferentes Marcas
group_anova01 = df_anova01.groupby(["Marca"])

#Usando el metodo f_oneway de scipy
resultado_anova = stats.f_oneway(group_anova01.get_group("honda")["Precio"], group_anova01.get_group("jaguar")["Precio"])
resultado_anova

#Nota: los precios de honda y jaguar son significativamente diferentes, ya que el puntaje de la Prueba F es mayor que 1 
#(muy GRANDE) y el Valor P es mayor que 0.05

#CONCLUSIONES: Podemos decir que hay una FUERTE correlacion entre una variable categorica y otras variables categoricas, si la
#prueba ANOVA nos da un valor de la Prueba F grande y un Valor P pequeño
```
